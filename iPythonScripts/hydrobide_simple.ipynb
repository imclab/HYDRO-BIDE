# This is the simplest hydrobide model. It does not include dormancy.## It includes resource limited growth, but does not include a parameter for cell maintenance. As a result, cell quota does not decrease; it can only increase. # Hydrobide: An individual-based community simulation of birth, immigration, death, emigration, resource limitation in a chemostat environment.There is 1 source, 1 local community, and 1 outlet. All BIDE processes occur via random sampling, i.e. individuals are picked at random to reproduce, immigrate, etc.The model is spatially implicit and individual-based. Information on every individual's Q and species identity is tracked. Resource concentration in the environment is also tracked. Some things are dynamic and will vary or arise naturally: Community size (N), composition, demography, cell quota (Q), resource concentration (R),nutrient uptake rate, biomass, biomass turnover. Others can remain constant across time: inflow rate (r), Volume (V). R influence resource uptake (v). v influences cell quota Q,and Q influences the per capita probability of reproduction. The rates of these processes are typiclally conceived to rely on the following:         GROWTH/REPRODUCTION: mu = mu_max*(1 - Qmin/Q)/(1 - Qmin/Qmax)        mu is growth rate        mu_max is maximal (nutrient saturated) growth rate        Q is the cell quota (nutrient per cell, cannot equal 0)        Qmin is the minimal cell quota needed for life                 UPTAKE: v/vmax = R/(K+R)  (basically monod)        v is the rate of uptake (what we want to know)        vmax = maximal rate of uptake        R is external resource concentration (which our model tracks)        K is the half saturation constant, i.e. value of R when v/vmax = 0.5

### Some diversity and evenness metrics

```python
# There are a lot of different metrics stashed in a 'metrics' module developed by KLdef e_var(SAD): # Smith and Wilson's evenness metric (Evar)     P = np.log(SAD)    S = len(SAD)    X = 0    for x in P:        X += (x - mean(P))**2/S    evar = 1 - 2/math.pi*np.arctan(X)     return(evar)def H(SAD): # Shannon's diversity    H = 0.0    N = sum(SAD)    for x in SAD:        x = x/float(N)        H += -1*(x*np.log(x))    return H
```

### Empirical constraints

```python
Vs = [100.0, 200.0, 400.0]  # list of Volume valuesR_cons = 0.4  # inflowing resource concentrationI_cons = 0.4  # propagule concentration of inflowing medium
```

### Theoretical / physiological constraints

```python
Qmin = 0.0 # min cell quotaK = 1.0 # half saturation constant
```

### Declare the number of time steps, initial number of steps to discard (burn-in), and lists to hold information to be plotted

# BEGIN THE SIMULATION

```python
for V in Vs: # FOR EACH VOLUME (V)        rs = range(int(V/20),int(V),int(V/20)) # list of flow rates    rs.reverse()        REStime = [] # list to hold residence times for a given volume        # Lists to hold time-averaged values for different residence times:    AvgNs = [] # mean total abundance    AvgRs = [] # mean resource density    AvgTRs = [] # mean total resources    AvgTOs = [] # mean biomass turnover    AvgSs = [] # mean richness    AvgHs = [] # mean Shannon's diversity    AvgEVARs = [] # mean evenness    AvgCTs = [] # mean compositional turnover    AvgAbs = [] # mean average abundance    AvgCTs = [] # mean compositional turnover    AvgQs = [] # mean cell quota        for r in rs: # FOR EACH RATE OF INFLOW (r)                Q = 0        I = int(round(I_cons * r)) # propagules immigrating per unit time          R = R_cons # initial resource concentration in the chemostat equals the inflowing concentration        TR = R*V   # total resources in the local environment        COM = []   # the community will be a list of lists                # Lists to hold values of total abundance, resource concentration, etc. across time for         # different combinations of volume (V) and inflow rate (r), i.e. residence times. Values        # in these lists will be averaged (i.e. across time after a burn-in period).        Nlist = [] # total abundance        Rlist = [] # resource concentration        TRlist = [] # total resources        TOlist = [] # biomass turnover        Slist = []  # richness        Evarlist = [] # evenness        Hlist = [] # Shannon's diversity        AvgAblist = [] # average abundance        CTlist = [] # compositional turnover        Qlist = [] # cell quota                SpeciesIn = [] # two lists used in finding compositional turnover        SpeciesOut = []        t = 0        while t <= time:             _in = I # records total number of immigrants (starts at I)            """ inflow of resources, Immigration of propagules """            lgp = 0.90 # arbitrarily set log-series parameter, though alpha is generally between 0.9 and < 1.0            propagules = np.random.logseries(lgp, I)  # An initial set of propagules; a list of log-series distributed integers.            # Assume the source community is infinite and that the regional community is log-series distributed.                        for propagule in propagules:                 Q = 0 # cell quota; Q = 0 means every propagule is dormant. Q can also be float(random.randrange(100))/100                # which makes cell quota (Q) of each propagule a uniform random variable between 0.0 and 1.0                                i = [propagule, Q] # an individual and its cell quota                COM.append(i)                N = len(COM) # size of the community                TR += r * R_cons # Total Resources increases due to inflow                R = TR/V # Resource concentration increases due to inflow                                        """ Reproduction/Growth/Death """            random.shuffle(COM) # randomize the community            for i, p in enumerate(COM):  # loop through the randomized community; one way to randomize individual changes                v = R/(K+R)   # amount of resource an individual uptakes                Q = p[1] + v  # increase cell quota by uptake                COM[i][1] = Q # change the individual's cell quota                TR -= v    # Total ambient resource supply decreases                R = TR/V   # producing a new ambient resource concentration                                      if Q >= 1.0: # reproduce if Q reaches or exceeds normalized Qmax, i.e. 1.0                    COM.append([p[0],Q/2.0]) # individuals produce sister cells with half Q                    COM[i][1] = Q/2.0                    _in += 1                               """ outflow of resources, Emigration """            out = int(round((N/V) * r)) # no. individuals lost per unit time            if out > 0:                while out > 0:                    i = randrange(0,len(COM)) # randomly pick an individual to die/emigrate                    SpeciesOut.append(COM[i][0]) # add to list of species removed                    COM.pop(i) # individual dies/emigrates                    out-=1 # number of lost             N = len(COM) # community size is reduced            SpeciesOut = list(set(SpeciesOut)) # makes list of species out non-redundant                        """ recording community info from time-steps """            if t >= burnin and t%10 == 0: # allow a burn-in and only record data every so many steps                                SpeciesIn = list(set(propagules)) # list of immigrating species                S_in = 0                SpeciesOut = list(set(SpeciesOut)) # makes list of species out non-redundant                S_out = 0                                for species in SpeciesIn:                    if species not in SpeciesOut: S_in += 1                for species in SpeciesOut:                    if species not in SpeciesIn: S_out += 1                                TQ = 0.0 # variable to hold total community cell quota                Slist1 = []                for i, p in enumerate(COM): # Find and record species richness                    Slist1.append(p[0])                    TQ += p[1]                Slist2 = list(set(Slist1))                S = len(Slist2)                                species_out = 0                for i, p in enumerate(Slist2): # Find number of species lost                    if p not in Slist1: species_out += 1                                   SAD = [] # Generate and save the rank-abundance distribution (RAD)                for i in Slist2:                    ab = Slist1.count(i)                    SAD.append(ab)                                SAD.sort()                SAD.reverse()                                # average values will be calculated from the lists below                if N > 0:                    Hlist.append(H(SAD))                    Evarlist.append(e_var(SAD))                    Nlist.append(np.log(N))                     Rlist.append(np.log(R))                    TRlist.append(TR)                    Slist.append(S)                    TOlist.append(np.log(abs(_in - out)))                    AvgAblist.append(N/S)                    CTlist.append(abs(S_in - S_out))                    Qlist.append(TQ/N)            t+=1                        # Append time-averaged values for total abundance, etc., to lists holding         # values for different residence times (but the same volume)         avgN = np.mean(Nlist)        if avgN > 0.0:            REStime.append(V/r)            AvgNs.append(avgN)                    avgR = np.mean(Rlist)            AvgRs.append(avgR)                    avgTR = np.mean(TRlist)            AvgTRs.append(avgTR)                    avgTO = np.mean(TOlist)            AvgTOs.append(avgTO)                    avgS = np.mean(Slist)            AvgSs.append(avgS)                    avgH = np.mean(Hlist)            AvgHs.append(avgH)                    avgEvar = np.mean(Evarlist)            AvgEVARs.append(avgEvar)                        avgAb = np.mean(AvgAblist)            AvgAbs.append(avgAb)                        avgCT = np.mean(CTlist)            AvgCTs.append(avgCT)                        avgTQ = np.mean(Qlist)            AvgQs.append(avgTQ)                        print 'V:',V,' res_time:',round(V/r,3),' N:',round(math.exp(avgN),2),' S:',round(avgS,2)             # Append lists holding time-averaged values for different residence times     # (but similar volumes) to lists holding values for different volumes:        Nlists.append(AvgNs)    Rlists.append(AvgRs)    TRlists.append(AvgTRs)    TOlists.append(AvgTOs)    REStimes.append(REStime)    Slists.append(AvgSs)    Hlists.append(AvgHs)    Evarlists.append(AvgEVARs)    AvgAblists.append(AvgAbs)    CTlists.append(AvgCTs)    Qlists.append(AvgQs)        print 'V:',V
```

### A function to generate plots of change with Residence Time

```python
def figSet(fig, i, Vs, Lists, REStimes, x_label, y_label):      ax = fig.add_subplot(1,3,i+1)    fs = 12 # fontsize    colors = ['b','r','c']    for j, _list in enumerate(Lists):          plt.plot(REStimes[j], _list, c=colors[j],label='Volume='+str(Vs[j]))    plt.xlabel(x_label, fontsize=fs)    plt.ylabel(y_label, fontsize=fs)    plt.xscale('log')    leg = plt.legend(loc=1, prop={'size':fs}) # legend specifics    leg.draw_frame(False) # no line around legend    plt.subplots_adjust(wspace=0.3, hspace=0.0)        return fig
```

### Generate plots of change in resources and abundance/biomass with increasing residence time 

```python
fig = plt.figure(figsize=(15.0,5.0))depVars = [Nlists,Rlists,TOlists] # dependent variablesy_labels = ['log(Biomass)','log(Resource density)','log(Biomass turnover)']x_label = 'Residence time'for i, depVar in enumerate(depVars):    figSet(fig, i, Vs, depVars[i], REStimes, x_label, y_labels[i])    plt.showplt.savefig('1-Rcons='+str(R_cons)+'-Icons='+str(I_cons)+'.png')
```
