# Hydrobide: An individual-based community simulation of birth, immigration, death, emigration, resource limitation in a chemostat environment.There is 1 source, 1 local community, and 1 outlet. All BIDE processes occur via random sampling, i.e. individuals are picked at random to reproduce, immigrate, etc.The model is spatially implicit and individual-based. Information on every individual's cell quota and species identity is tracked. Several aspects of the community are tracked.Some things are dynamic and will vary or arise naturally: Community size (N), composition, demography, cell quota (Q), resource concentration (R),nutrient uptake rate, biomass, biomass turnover. Others can remain constant across time: inflow rate (r), Volume (V). R influence resource uptake (v). v influences cell quota Q,and Q influences the per capita probability of reproduction. The rates of these processes are typiclally conceived to rely on the following:     ### GROWTH/REPRODUCTION: mu = mu_max*(1 - Qmin/Q)/(1 - Qmin/Qmax)mu: growth ratemu_max: maximal (nutrient saturated) growth rateQ: cell quota (nutrient per cell, cannot equal 0)Qmin: minimal cell quota needed for active metabolism (Q at or below Qmin implies death or dormancy)             ### UPTAKE: v/vmax = R/(K+R)  (basically monod) v: rate of uptake (what we want to know)vmax: maximal rate of uptakeR: external resource concentration (which our model tracks)K: half saturation constant, i.e. value of R when v/vmax = 0.5 (for simplifying purposes K = 1)

### Import some modules

```python
import sys                                            import  matplotlib.pyplot as pltimport numpy as np                                    from scipy import stats                                 import random                                           from random import choice, randrangeimport math
```

### Some diversity and evenness metrics

```python
# There are a lot of different metrics stashed in a 'metrics' module developed by KLdef e_var(SAD): # Smith and Wilson's evenness metric (Evar)     P = np.log(SAD)    S = len(SAD)    X = 0    for x in P:        X += (x - mean(P))**2/S    evar = 1 - 2/math.pi*np.arctan(X)     return(evar)def H(SAD): # Shannon's diversity    H = 0.0    N = sum(SAD)    for x in SAD:        x = x/float(N)        H += -1*(x*np.log(x))    return H
```

### Empirical constraints

```python
Vs = [10.0, 50.0, 100.0]  # list of Volume valuesR_cons = 3.0  # inflowing resource concentrationI_cons = 3.0  # propagule concentration of inflowing medium
```

### Theoretical physiological constraint(s)

```python
maintE = 0.1 # maintenance energy for active metabolism
```

### Declare the number of time steps, initial number of steps to discard (burn-in), and lists to hold information to be plotted

```python
time = 500  # length of the experimentburnin = int(0.5*time)  # number of initial time steps to discard# Each of the following is a list of lists, each sublist holding values# for different residence times but similar volumesNlists = [] # mean total abundanceRlists = [] # mean resource concentrationTRlists = [] # mean total resourcesTOlists = [] # mean biomass TurnoverREStimes = [] # residence times for different VolumesSlists = [] # mean richness Hlists = [] # mean Shannon's diversityEvarlists = [] # mean evenness (Evar)AvgAblists = [] # mean average abundanceCTlists = [] # mean compositional turnoverQlists = [] # mean cell quotaDormlists = [] # mean relative dormant abundance
```

# BEGIN THE SIMULATION

```python
for V in Vs: # FOR EACH VOLUME (V)        rs = [] # list of flow rates    divisor = 1.0    step = V/20.0    while divisor <= V:        rs.append(V/divisor)        divisor += step            REStime = [] # list to hold residence times for a given volume    # Lists to hold time-averaged values for different residence times:    AvgNs = [] # mean total abundance    AvgRs = [] # mean resource density    AvgTRs = [] # mean total resources    AvgTOs = [] # mean biomass turnover    AvgSs = [] # mean richness    AvgHs = [] # mean Shannon's diversity    AvgEVARs = [] # mean evenness    AvgCTs = [] # mean compositional turnover    AvgAbs = [] # mean average abundance    AvgCTs = [] # mean compositional turnover    AvgQs = [] # mean cell quota    AvgDorms = [] # mean relative dormant abundance        for r in rs: # FOR EACH RATE OF INFLOW (r)                Q = 0        I = int(round(I_cons * r)) # propagules immigrating per unit time          R = R_cons # initial resource concentration in the chemostat equals the inflowing concentration        TR = R*V   # total resources in the local environment        COM = []   # the community will be a list of lists                # Lists to hold values of total abundance, resource concentration, etc. across time for         # different combinations of volume (V) and inflow rate (r), i.e. residence times. Values        # in these lists will be averaged (i.e. across time after a burn-in period).        Nlist = [] # total abundance        Rlist = [] # resource concentration        TRlist = [] # total resources        TOlist = [] # biomass turnover        Slist = []  # richness        Evarlist = [] # evenness        Hlist = [] # Shannon's diversity        AvgAblist = [] # average abundance        CTlist = [] # compositional turnover        Qlist = [] # cell quota        Dormlist = [] # relative dormant abundance                """ initial inflow of resources, Immigration of propagules """        lgp = 0.90 # arbitrarily set log-series parameter, though alpha is generally between 0.9 and < 1.0        propagules = np.random.logseries(lgp, I)  # An initial set of propagules; a list of log-series distributed integers.        # Assume the source community is infinite and that the regional community is log-series distributed.                for propagule in propagules:             Q = 0 # cell quota; Q = 0 means every propagule is dormant. Q can also be float(random.randrange(100))/100            # which makes cell quota (Q) of each propagule a uniform random variable between 0.0 and 1.0            i = [propagule, Q] # an individual and its cell quota            COM.append(i)            TR += r * R_cons # Total Resources increases due to inflow            R = TR/V # Resource concentration increases due to inflow                S1 = list(set(propagules)) # list of initial immigrating species (the starting community)        N1 = len(COM)                t = 0        while t <= time:                        """ Growth, Reproduction """            random.shuffle(COM) # randomize the community            daughterCells = []            for i, ind in enumerate (COM):                v = R/(1+R)   # amount of resource an individual takes up                Q = ind[1] + v  # increase cell quota by uptake                if Q > 0.0: Q -= maintE # decrease cell quota by maintenance energy                COM[i][1] = Q # change the individual's cell quota                TR -= v    # Total ambient resource supply decreases                R = TR/V   # producing a new ambient resource concentration                              if Q >= 1.0: # reproduce if Q reaches or exceeds normalized Qmax, i.e. 1.0                    daughterCells.append([ind[0],Q/2.0]) # individuals produce sister cells with half Q                    COM[i][1] = Q/2.0                                COM.extend(daughterCells)            """ outflow of resources, Emigration/Death """            out = int(round((N1/V) * r)) # no. individuals lost per unit time            if out > 0:                while out > 0:                    i = randrange(0,len(COM)) # randomly pick an individual to die/emigrate                    COM.pop(i) # individual dies/emigrates                    out-=1                                """ recording community info from time-steps """            if t >= burnin and t%10 == 0: # allow a burn-in and only record data every so many steps                                dorms = 0 # number of dormant individuals                TQ = 0.0 # variable to hold total community cell quota                S = []                for i, p in enumerate(COM): # Find and record species richness and total cell quota                    S.append(p[0])                    q = p[1]                    TQ += q                    if q <= 0.0:                        dorms += 1                                N2 = len(COM) # community size is reduced                S2 = list(set(S))                """ Beta diversity (turnover): Ba = (S1 - c) + (S2 - c)                James S. Albert; Roberto E. Reis (8 March 2011). Historical Biogeography of                Neotropical Freshwater Fishes. University of California Press. p. 308."""                c = len(list(set(S1) & set(S2)))                s1 = len(S1)                s2 = len(S2)                Ba = (s1 - c) + (s2 - c)                                #print 'Ba',Ba,'s1',s1,'s2',s2,'c',c                                RAD = [] # Generate and save the rank-abundance distribution (RAD)                for i in S2:                    ab = S.count(i)                    RAD.append(ab)                                RAD.sort()                RAD.reverse()                                # average values will be calculated from the lists below                if N2 > 0:                    #print 'Ba',Ba,'s1',s1,'s2',s2,'c',c                    N2 = float(N2)                    S2 = float(s2)                    Hlist.append(H(RAD))                    Evarlist.append(e_var(RAD))                    Nlist.append(np.log(N2))                     Rlist.append(np.log(R))                    TRlist.append(TR)                    Slist.append(S2)                    TOlist.append(abs(N1 - N2))                    AvgAblist.append(np.log(N2/S2))                    CTlist.append(Ba)                    Qlist.append((TQ/N2)*S2)                    Dormlist.append(dorms/N2)                                    """ inflow of resources, Immigration of propagules """            lgp = 0.90 # arbitrarily set log-series parameter, though alpha is generally between 0.9 and < 1.0            propagules = np.random.logseries(lgp, I)  # An initial set of propagules; a list of log-series distributed integers.            # Assume the source community is infinite and that the regional community is log-series distributed.            SpIm = list(set(propagules)) # list of immigrating species                        for propagule in propagules:                 Q = 0 # cell quota; Q = 0 means every propagule is dormant. Q can also be float(random.randrange(100))/100                # which makes cell quota (Q) of each propagule a uniform random variable between 0.0 and 1.0                                i = [propagule, Q] # an individual and its cell quota                COM.append(i)                N = len(COM) # size of the community                TR += r * R_cons # Total Resources increases due to inflow                R = TR/V # Resource concentration increases due to inflow                        N1 = len(COM)            S1 = list(set(SpIm) | set(S1))             t+=1                        # Append time-averaged values for total abundance, etc., to lists holding         # values for different residence times (but the same volume)         avgN = np.mean(Nlist)        if avgN > 0.0:            REStime.append(1/r)            AvgNs.append(avgN)                    avgR = np.mean(Rlist)            AvgRs.append(avgR)                    avgTR = np.mean(TRlist)            AvgTRs.append(avgTR)                    avgTO = np.mean(TOlist)            AvgTOs.append(avgTO)                    avgS = np.mean(Slist)            AvgSs.append(avgS)                    avgH = np.mean(Hlist)            AvgHs.append(avgH)                    avgEvar = np.mean(Evarlist)            AvgEVARs.append(avgEvar)                        avgAb = np.mean(AvgAblist)            AvgAbs.append(avgAb)                        avgCT = np.mean(CTlist)            AvgCTs.append(avgCT)                        avgTQ = np.mean(Qlist)            AvgQs.append(avgTQ)                        avgDorm = np.mean(Dormlist)            AvgDorms.append(avgDorm)                        print 'V:',V,' res_time:',round(V/r,3),' N:',round(math.exp(avgN),2),' S:',round(avgS,2),' Seed bank size:',round(avgDorm,3)             # Append lists holding time-averaged values for different residence times     # (but similar volumes) to lists holding values for different volumes:        Nlists.append(AvgNs)    Rlists.append(AvgRs)    TRlists.append(AvgTRs)    TOlists.append(AvgTOs)    REStimes.append(REStime)    Slists.append(AvgSs)    Hlists.append(AvgHs)    Evarlists.append(AvgEVARs)    AvgAblists.append(AvgAbs)    CTlists.append(AvgCTs)    Qlists.append(AvgQs)    Dormlists.append(AvgDorms)        print 'V:',V
```

### A function to generate plots of change with Residence Time

```python
def figSet(fig, i, Vs, Lists, REStimes, x_label, y_label):      ax = fig.add_subplot(1,3,i+1)    fs = 12 # fontsize    colors = ['b','r','c']    if i == 0:        for j, _list in enumerate(Lists):              plt.plot(REStimes[j], _list, c=colors[j],label='Volume='+str(Vs[j]))            leg = plt.legend(loc=1, prop={'size':fs}) # legend specifics            leg.draw_frame(False) # no line around legend    elif i > 0:        for j, _list in enumerate(Lists):              plt.plot(REStimes[j], _list, c=colors[j])    plt.xlabel(x_label, fontsize=fs)    plt.ylabel(y_label, fontsize=fs)    plt.subplots_adjust(wspace=0.3, hspace=0.0)        return fig
```

### Generate plots of biomass, percent biomass turnover, & percent turnover in composition with increasing residence time 

```python
fig = plt.figure(figsize=(15.0,5.0))depVars = [Nlists, TOlists, CTlists] # dependent variablesy_labels = ['ln(Biomass)','Biomass turnover','Beta diversity']x_label = 'Normalized Residence time, (1/r)'for i, depVar in enumerate(depVars):    figSet(fig, i, Vs, depVars[i], REStimes, x_label, y_labels[i])    plt.showplt.savefig('1-Rcons='+str(R_cons)+'-Icons='+str(I_cons)+'.png')
```
